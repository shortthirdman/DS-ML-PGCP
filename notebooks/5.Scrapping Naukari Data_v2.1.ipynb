```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import pandas as pd
```

```python
url = "https://www.naukri.com/financial-analyst-jobs-in-mumbai"
```

```python
driver = webdriver.Chrome("C:\\Users\\Raghavendra N\\OneDrive\\Official\\ChromeDriver\\chromedriver.exe")
```

```python
driver.get(url)
time.sleep(10)
soup = BeautifulSoup(driver.page_source,'html5lib')
```

```python
print(soup.prettify())
```

```python
driver.close()
```

```python
results = soup.find(class_='list')
```

```python
job_elem = results.find('article',class_='jobTuple bgWhite br4 mb-8')
```

```python
job_url = job_elem.find('a',class_='title fw500 ellipsis').get('href')
job_url
```

```python
Title = job_elem.find('a',class_='title fw500 ellipsis').get('title')
Title
```

```python
Company = job_elem.find('a',class_='subTitle ellipsis fleft').get('title')
Company
```

```python
rating = job_elem.find('span',class_='starRating fleft dot')
if rating is None:
    rating = None
else:
    rating = rating.text
print(rating)
```

```python
Review_counts = job_elem.find('a',class_='reviewsCount ml-5 fleft blue-text')
if Review_counts is None:
    Review_counts = None
else:
    Review_counts = Review_counts.text
print(Review_counts)
```

```python
Exp_root = job_elem.find('li',class_='fleft grey-text br2 placeHolderLi experience')
Experience = Exp_root.find('span',class_='ellipsis fleft fs12 lh16')
if Experience is None:
    Experience = None
else:
    Experience = Experience.get('title')
print(Experience)
```

```python
sal_root = job_elem.find('li',class_='fleft grey-text br2 placeHolderLi salary')
Salary = sal_root.find('span',class_='ellipsis fleft fs12 lh16')
if Salary is None:
    Salary = None
else:
    Salary = Salary.get('title')
print(Salary)
```

```python
location_root = job_elem.find('li',class_='fleft grey-text br2 placeHolderLi location')
Location = location_root.find('span',class_='ellipsis fleft fs12 lh16')
if Location is None:
    Location = None
else:
    Location = Location.get('title')
print(Location)
```

```python
df = pd.DataFrame(columns=['Title','Company','Ratings','Reviews','Experience','Salary','Location','URL'])
```

```python
df=df.append({'URL':job_url,'Title':Title,'Company':Company,'Ratings':rating,'Reviews':Review_counts,
              'Experience':Experience,'Salary':Salary,'Location':Location},ignore_index = True)
```

```python
df = pd.DataFrame(columns=['Title','Company','Ratings','Reviews','Experience','Salary','Location','URL'])
results = soup.find(class_='list')
job_elems = results.find_all('article',class_='jobTuple bgWhite br4 mb-8')
for job_elem in job_elems:
    job_url = job_elem.find('a',class_='title fw500 ellipsis').get('href')
    Title = job_elem.find('a',class_='title fw500 ellipsis').get('title')
    Company = job_elem.find('a',class_='subTitle ellipsis fleft').get('title')
    rating = job_elem.find('span',class_='starRating fleft dot')
    if rating is None:
        rating = None
    else:
        rating = rating.text
    Review_counts = job_elem.find('a',class_='reviewsCount ml-5 fleft blue-text')
    if Review_counts is None:
        Review_counts = None
    else:
        Review_counts = Review_counts.text
    Exp = job_elem.find('li',class_='fleft grey-text br2 placeHolderLi experience')
    Experience = Exp.find('span',class_='ellipsis fleft fs12 lh16')
    if Experience is None:
        Experience = None
    else:
        Experience = Experience.get('title')
    sal_root = job_elem.find('li',class_='fleft grey-text br2 placeHolderLi salary')
    Salary = sal_root.find('span',class_='ellipsis fleft fs12 lh16')
    if Salary is None:
        Salary = None
    else:
        Salary = Salary.get('title')
    location_root = job_elem.find('li',class_='fleft grey-text br2 placeHolderLi location')
    Location = location_root.find('span',class_='ellipsis fleft fs12 lh16')
    if Location is None:
        Location = None
    else:
        Location = Location.get('title')
    df=df.append({'URL':job_url,'Title':Title,'Company':Company,'Ratings':rating,'Reviews':Review_counts,
              'Experience':Experience,'Salary':Salary,'Location':Location},ignore_index = True)
```
